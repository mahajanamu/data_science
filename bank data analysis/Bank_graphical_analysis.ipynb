```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv("data/bank-full.csv",sep=";")
data.head()
```

### Question 1
Question1
From the given dataset create line plots to visualize the number of campaigns for the given months. Here perform the dot plots for the months April, May and June.

• Set figure height as 5 inches and figure width as 15 inches.

• Set the font size of the main title as 15.

• Set the font size of the x-axis and y-axis label as 12.

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(111)
sel_month = data[data["month"].isin(["may","apr","jun"])]
sns.lineplot(data=sel_month,x="campaign",y="day",hue="month",style="month",style_order=["may","apr","jun"],hue_order=["may","apr","jun"],markers=["o","o","o"],dashes=False,ax=ax,legend=False)
sns.lineplot(data=sel_month,x="campaign",y="day",hue="month",hue_order=["may","apr","jun"],ax=ax,legend="brief")
ax.set_title("Number of campaigns based in Months",fontsize=15,fontweight="bold")

ax.set_xlabel("Campaigns",fontsize=12)
ax.set_ylabel('Day',fontsize=12)
sns.set_style("whitegrid")
plt.xticks(rotation=45)
plt.show()
```

### Question 3
From the given dataset create a bar plot to visualize the count of customers with respect to the personal and housing loan based on each job category.

• Set figure height as 5 inches and figure width as 10 inches.

• Set the font size of the main title as 15.

• Set the font size of the x-axis and y-axis label as 12.


```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(111)
h_yes={}
h_no={}
i_yes={}
i_no = {}
for name in list(data["job"].unique()):
    h_yes[name]=len(data[(data["job"]==name) & (data["housing"]=="yes")])
    h_no[name]=len(data[(data["job"]==name) & (data["housing"]=="no")])
    i_yes[name]=len(data[(data["job"]==name) & (data["loan"]=="yes")])
    i_no[name]=len(data[(data["job"]==name) & (data["loan"]=="no")])
X_axis = np.arange(len(h_yes))
plt.bar(X_axis,list(h_yes.values()),width=0.2,color="dodgerblue")
plt.bar(X_axis+0.2,list(h_no.values()),width=0.2,color="orange")
plt.bar(X_axis+0.4,list(i_yes.values()),width=0.2,color="green")
plt.bar(X_axis+0.6,list(i_no.values()),width=0.2,color="red")
plt.xticks(X_axis + 0.3,list(h_yes.keys()))
plt.legend(labels=["h_yes","h_no","I_yes","I_no"])
plt.xlim(-.5,len(X_axis))
plt.ylim(0,8500)
ax.set_title("Job and loan",fontsize=15,fontweight="bold")
ax.set_xlabel("jobs",fontsize=12)
ax.set_ylabel('Count',fontsize=12)
plt.xticks(rotation=90)
plt.show()
```

### Question 4
From the given dataset create dot plots to visualize the number of campaigns with respect to the duration for the given months. Here perform the dot plots for the months January, February, March, April and May.

• Set figure height and figure width as 10 inches.

• Set the font size of the main title as 15.

• Set the font size of the x-axis and y-axis label as 12.



```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)
req_mon = ["jan","feb","mar","apr","may"]
req_data = data[data["month"].isin(req_mon)]
ptplot = sns.scatterplot(data=req_data,x="campaign",y="duration",hue="month",hue_order=req_mon,palette=["green","orange","darkred","blue","red"])
ax.set_title(" Campaigns Vs Duration - Month Wise ",fontsize=15,fontweight="bold")
ax.set_xlabel(" Campaigns ",fontsize=12)
ax.set_ylabel(' Duration (Second) ',fontsize=12)
sns.set_style("white")
```
